```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/Salary_Data.csv")
x = dataset.iloc[:, :-1].values #ozellikler / bagimsiz degisken
y = dataset.iloc[:, -1].values #ozellikler / bagimli degisken
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

#Basit Lineer Regresyon modelinin Ogrenme verisinde ogrenmesinin gerceklestirmesi

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(xTrain, yTrain) # ogrenme asamasi
```

#Test verisi üzerinde tahminlerin yapılması


```python
yPred = regressor.predict(xTest)
```

#Ogrenme sonuclarinin gorsellestirilmesi

```python
plt.scatter(xTrain, yTrain, color = 'red')
plt.plot(xTrain, regressor.predict(xTrain), color = 'blue')
plt.title('Maas ve Deneyim (Ogrenme Verisi)')
plt.xlabel('Deneyim')
plt.ylabel('Maas')
plt.show()
```

#Test sonuclarinin gorsellestirilmesi

```python
plt.scatter(xTest, yTest, color = 'red')
plt.plot(xTrain, regressor.predict(xTrain), color = 'blue')
plt.title('Maas ve Deneyim (Ogrenme Verisi)')
plt.xlabel('Deneyim')
plt.ylabel('Maas')
plt.show()
```
