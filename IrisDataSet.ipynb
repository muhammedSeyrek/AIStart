```python
from sklearn.datasets import load_iris
irisDataSet = load_iris()
print(irisDataSet)
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(irisDataSet['data'], irisDataSet['target'], test_size = 0.20)
print(xTrain.shape)
print(xTest.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kNN = KNeighborsClassifier(n_neighbors = 1)
print(kNN)
```

```python
kNN.fit(xTrain, yTrain)
```

```python
xNew = [[3.5, 2.1, 3.4, 1.2]]
guess = kNN.predict(xNew)
```

```python
accuracy = kNN.predict(xTest)
print(accuracy)
```

```python
import numpy as np
np.mean(accuracy == yTest) * 100
```
